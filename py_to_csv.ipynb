```python
import pandas as pd
from metro_stations import metro_stations

# Convert line-station mapping to a dataframe
rows = []
for line, stations in metro_stations.items():
    if line != 'interchange_stations':  # handle separately
        for station in stations:
            rows.append({'Metro Line': line, 'Station Name': station})

# Convert interchange stations too
for station, lines in metro_stations['interchange_stations'].items():
    for line in lines:
        rows.append({'Metro Line': line, 'Station Name': station, 'Is Interchange': True})

# Create dataframe
station_df = pd.DataFrame(rows)

# Fill missing 'Is Interchange' with False
station_df['Is Interchange'] = station_df.get('Is Interchange', False)

# Save to CSV
station_df.to_csv('bangalore_metro_stations.csv', index=False)
```
